Calculate VaR

2. Using a normal distribution with an Exponentially Weighted variance (Œª = 0. 94) 
3. Using a MLE fitted T distribution. 
4. Using a fitted AR(1) model.

```python
df=pd.read_csv('DailyPrices.csv')
df
```

```python
df.drop('Date', axis=1, inplace=True)
rt=df.pct_change().dropna()
```

```python
rt_mean=rt.mean()
rt1=rt-rt_mean
rt=rt1.loc[:,'META']
rt
```

```python
#Using normal distribution
confidence_level = 0.95
z_alpha = stats.norm.ppf(1 - confidence_level)
var_normal=-(np.sqrt(np.var(rt)) * z_alpha)
print("VaR normal distribution: {}".format(var_normal))
```

```python
# Using normal distribution with EWMA variance
ewma_lambda = 0.94
column=pd.DataFrame(column, index=None)
z_alpha = stats.norm.ppf(1 - confidence_level)
ewma_var = rt.ewm(alpha=1-ewma_lambda).var().iloc[-1]
var_ewma=-(float(np.sqrt(ewma_var) * z_alpha))
var_ewma
print("VaR normal distribution with EWMA variance: {}".format(var_ewma))
```

```python
# Using MLE fitted T distribution
t_params = stats.t.fit(rt)
var_t=-stats.t.ppf(0.05, df=t_params[0], loc=t_params[1], scale=t_params[2])
print("VaR MLE fited T distribution: {}".format(var_t))
```

```python
# Using fitted AR(1) model
from statsmodels.tsa.arima.model import ARIMA
n=10000
model = ARIMA(rt, order=(1, 0, 0)).fit()
sigma = np.std(model.resid)
sim_returns = np.empty(n)
returns = rt.values
for i in range(n):
    sim_returns[i] =  model.params[0] * (returns[-1]) + sigma * np.random.normal()
var_AR1 = -np.percentile(sim_returns, 0.05*100)
print("VaR AR(1): {}".format(var_AR1))
```

```python
# Using historic simulation
num_simulations = 1000
simulated_returns = np.random.choice(rt.squeeze(), size=(num_simulations, len(rt)))
var_historic=(-np.percentile(simulated_returns, (1-confidence_level)*100))
print("VaR historical simulation: {}".format(var_historic))
```
